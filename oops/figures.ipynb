```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
x = np.random.sample(100)
```

```python
x = np.random.random_sample(50)
y = 0.45 * x

xx = np.random.random_sample(50)
z = (1 - 0.3) * xx + 0.3

aa = np.concatenate((y, z))
```

```python
plt.figure(figsize=(8, 3))
plt.plot(aa, color='orange')
plt.xlabel('Time', fontsize = 16)
plt.ylabel('Probabilities', fontsize = 16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.savefig('teaser.png', dpi=500)
```

```python
x = np.random.random_sample(45)
y = 0.45 * x

xx = np.random.random_sample(60)
z = (1 - 0.4) * xx + 0.4

w = np.linspace(0, 1, 5)

bb = np.concatenate((y, z))
y = pd.Series(y).rolling(10).mean().iloc[10:]
z = pd.Series(z).rolling(10).mean().iloc[10:]

inter = (1 - w) * y.iloc[-1] + w * z.iloc[0]

inter = pd.Series(inter)

bb = np.concatenate((y, inter, z))
```

```python

fig, ax1 = plt.subplots(figsize=(10, 5))

color = 'tab:blue'
ax1.set_xlabel('Time', fontsize=18)
ax1.set_ylabel('Probabilities', color=color, fontsize=18)
ax1.plot(aa, color=color, alpha=0.6)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_ylim(0, 1.2)

#ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:orange'
#ax2.set_ylabel('Real signal', color=color, fontsize=18)  # we already handled the x-label with ax1
ax2.plot(bb, color=color)
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_ylim(0, 1.2)

ax1.vlines(39, 0.05, 1., color='green')
ax1.text(17, 0.7, 'Real Change', color='green', fontsize=14)
ax1.text(24, 0.6, 'Point', color='green', fontsize=14)

ax1.vlines(50, 0.05, 1., color='black', linestyle='dashed', alpha=0.7)
ax1.text(51, 0.2, 'Predicted Change', color='black', fontsize=14)
ax1.text(60, 0.1, 'Point', color='black', fontsize=14)

#ax1.arrow(39, 1, 9, 0, width=0.01, head_length=2)
ax1.annotate(s='', xy=(39,1), xytext=(50,1), 
             arrowprops=dict(arrowstyle='<->', color='brown'))
ax1.text(40, 1.05, 'Detection Delay', color='brown', fontsize=16)


fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.savefig('delay.png', dpi=500, bbox_inches='tight')
```

```python

fig, ax1 = plt.subplots(figsize=(10, 5))

color = 'tab:blue'
ax1.set_xlabel('Time', fontsize=18)
ax1.set_ylabel('Probabilities', color=color, fontsize=18)
ax1.plot(aa, color=color, alpha=0.6)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_ylim(0, 1.2)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:orange'
ax2.set_ylabel('Real signal. No change', color=color, fontsize=18)  # we already handled the x-label with ax1
ax2.plot(bb, color=color)
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_ylim(0, 1.2)

ax1.vlines(50, 0.05, 1., color='black', linestyle='dashed', alpha=0.7)
ax1.text(26, 0.6, 'Predicted Change', color='black', fontsize=14)
ax1.text(34, 0.5, 'Point', color='black', fontsize=14)

ax1.annotate(s='', xy=(0,1), xytext=(50,1), 
             arrowprops=dict(arrowstyle='<->', color='brown'))
ax1.text(10, 1.05, 'False Positive Delay', color='brown', fontsize=16)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.savefig('fp_delay.png', dpi=500, bbox_inches='tight')
```

```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd
```

```python
final_res_alpha = pd.read_csv('compare_alphas_10_runs.csv')
```

```python
final_res_alpha
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(data=final_res_alpha[final_res_alpha['0.0'] < 500], width=0.5, fliersize=0, color='#9fbbcf')
sns.swarmplot(data=final_res_alpha[final_res_alpha['0.0'] < 500], color='k', alpha=0.5)
plt.xticks(fontsize=22);
plt.yticks(fontsize=22);
plt.xlabel('Weight of Delay Loss', fontsize=26)
plt.ylabel('Result AUC', fontsize=26);
plt.grid(axis='y')
plt.savefig('compare_alphas.pdf', bbox_inches='tight')
```
